```python
filtered_india_df = pandas_df[pandas_df['country'] == 'india']
```

```python
filtered_india_df.count()
```

```python
filtered_SEAR_df = pandas_df[pandas_df['who_region'] == 'sear']
```

```python
filtered_SEAR_df.count()
```

```python
#Trend Analysis: Visualize the daily new cases and new deaths over time globally or for specific countries.
```

```python
fig = px.scatter(filtered_india_df, x = "date_reported", y="new_cases", title="new_cases_over_day_for_India")
fig.show()
```

```python
fig = px.scatter(filtered_india_df, x = 'date_reported', y ='new_deaths', title="new_daily_deaths_in_india")
fig.show()
```

```python
fig = px.scatter(filtered_SEAR_df, x = "date_reported", y ="new_cases", color='country', title="new_Cases in SEAR")
fig.show()
```

```python
fig = px.scatter(filtered_SEAR_df, x="date_reported", y="new_deaths", color="country", title="new deaths in SEAR")
fig.show()
```

```python
#Regional Comparison: Compare the cumulative cases and cumulative deaths across WHO regions using bar or pie charts.
```

```python
max_deaths_SEAR_df = filtered_SEAR_df.groupby('country', as_index=False).max('cumulative_deaths')
```

```python
max_deaths_SEAR_df
```

```python
fig = px.pie(max_deaths_SEAR_df, values='cumulative_deaths', names='country', title="cumulative_deaths for each country in SEAR")
fig.show()
```

```python
max_cases_SEAR_df = filtered_SEAR_df.groupby('country', as_index=False).max('cumulative_cases')
```

```python
max_cases_SEAR_df
```

```python
fig = px.pie(max_cases_SEAR_df, values='cumulative_cases', names='country', title="cumulative cases in SEAR")
fig.show()
```

```python
#Country-Level Drilldown: Show the top 10 countries with the highest cumulative cases and deaths.
```

```python
max_cumulative_pandas_df = pandas_df.groupby("country", as_index=False).max('cumulative_cases')
max_cumulative_pandas_df = max_cumulative_pandas_df.sort_values('cumulative_cases', ascending=False).head(10)
max_cumulative_pandas_df
```

```python
fig = px.bar(max_cumulative_pandas_df, x='country', y='cumulative_cases', title='most cases country wise' )
fig.show()
```

```python
max_cumulative_deaths_df = pandas_df.groupby('country', as_index= False).max('cumulative_death')
max_cumulative_deaths_df = max_cumulative_deaths_df.sort_values('cumulative_deaths', ascending= False).head(10)
max_cumulative_deaths_df
```

```python
fig = px.bar(max_cumulative_deaths_df, x='country', y='cumulative_deaths', color='country', title="max_deaths_country_wise")
fig.show()
```
